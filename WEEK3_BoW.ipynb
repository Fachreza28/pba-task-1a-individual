#BoW bag of Words


# Step 1: Tokenize the data, remove stop words and perform stemming or lemmatization.

```python
!pip install sastrawi --quiet

import nltk
nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('wordnet')

import re
import pandas as pd
from collections import Counter
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#### Load CSV dari local / upload
from google.colab import files
uploaded = files.upload()  # upload CSV dari local

df = pd.read_csv('df_mybcarev_after_stopwords_sentiment_rating2.csv')  # sesuaikan path-nya
print(df.shape)
df.head()
```

```python
# Setup stemmer Sastrawi + stopwords Indonesia
factory = StemmerFactory()
stemmer = factory.create_stemmer()

indonesian_stopwords = set(stopwords.words('indonesian'))

def preprocess(text):
    if not isinstance(text, str):
        return ''
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = text.lower()
    tokens = text.split()
    tokens = [word for word in tokens if word not in indonesian_stopwords and len(word) > 2]
    # Stem seluruh kalimat sekaligus (bukan per kata)
    stemmed = stemmer.stem(' '.join(tokens))
    return stemmed

df['processed'] = df['content'].apply(preprocess)

corpus = df['processed'].dropna().tolist()
corpus = [doc for doc in corpus if doc.strip() != '']

print("Jumlah dokumen:", len(corpus))
print("Contoh hasil preprocessing:")
print(corpus[:3])
```

# Step 2: List all unique words

```python
all_words = []
for sentence in corpus:
    for word in sentence.split():
        all_words.append(word)

unique_words = list(set(all_words))
print(f"Jumlah unique words: {len(unique_words)}")
print("Contoh unique words:", unique_words[:20])
```

```python
word_frequencies = Counter(all_words)

# Sorting by frequency descending
sorted_words = sorted(word_frequencies.items(), key=lambda item: item[1], reverse=True)

# Mapping word ke nomor
word_to_number = {word: i + 1 for i, (word, freq) in enumerate(sorted_words)}

print("Top 20 kata paling sering muncul:")
for word, freq in sorted_words[:20]:
    print(f"  '{word}' → frekuensi: {freq}, nomor: {word_to_number[word]}")
```

```python
word_frequencies = Counter(all_words)

# Sorting by frequency descending
sorted_words = sorted(word_frequencies.items(), key=lambda item: item[1], reverse=True)

# Mapping word ke nomor
word_to_number = {word: i + 1 for i, (word, freq) in enumerate(sorted_words)}

print("Top 20 kata paling sering muncul:")
for word, freq in sorted_words[:20]:
    print(f"  '{word}' → frekuensi: {freq}, nomor: {word_to_number[word]}")
```

```python
# Batasi vocabulary agar matrix tidak terlalu besar
cv = CountVectorizer(max_features=1000)  # ambil 1000 kata teratas
bow_matrix = cv.fit_transform(corpus).toarray()

print("Shape BoW matrix:", bow_matrix.shape)
print("Contoh vocabulary (20 pertama):", cv.get_feature_names_out()[:20])
print("\nContoh baris pertama BoW:\n", bow_matrix[0])
```

```python
bow_df = pd.DataFrame(bow_matrix, columns=cv.get_feature_names_out())
print(bow_df.shape)
bow_df.head()
```
