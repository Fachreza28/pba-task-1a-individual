```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

```

```python
# Input Data (Berita)
documents = [
    """Kebakaran terjadi di area kilang Pertamina Balongan. Api muncul dari area mesin dan menyebabkan kepanikan warga.""",
    """Api berhasil dipadamkan oleh petugas setelah kebakaran terjadi di kilang minyak Pertamina.""",
    """Warga diungsikan akibat kebakaran hebat di area Pertamina Balongan."""
]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)
```

```python
# 3. Hitung Cosine Similarity
# Membandingkan semua dokumen dengan Dokumen 0 (1 John)
cos_sim = cosine_similarity(X[0:1], X)

# Tampilkan Hasil
df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
df['Cosine Sim to Doc 0'] = cos_sim.flatten()
print(df)
```

**CountVectorizer:** Mengubah teks menjadi angka (matriks frekuensi kata). Di sini kita membatasi fitur hanya pada kata "hate" dan "darkness" agar sesuai dengan grafik di slide Stanford.

**Cosine Similarity vs Distance:** Fungsi ini menghasilkan nilai 0 sampai 1. Semakin mendekati 1, semakin mirip kedua dokumen tersebut.

**Analisis Hasil:** Bahwa Dokumen 2 (MLK) memiliki nilai similarity yang tinggi terhadap Dokumen 0, membuktikan bahwa meskipun jumlah katanya jauh lebih banyak, "arah" topiknya tetap serupa.

**Euclidean Distance:** Jika kita menggunakan jarak lurus, Dokumen 2 akan dianggap sangat berbeda karena jumlah kemunculan katanya (3 dan 4) jauh dari Dokumen 0 (1 dan 2).

**Aplikasi di Dunia Nyata:** Teknik ini adalah dasar dari mesin pencari (search engines) dan sistem rekomendasi teks yang sering kita gunakan sehari-hari.

```python
documents = [
    """Kebakaran terjadi di area kilang Pertamina Balongan. Api muncul dari area mesin dan menyebabkan kepanikan warga.""",
    """Api berhasil dipadamkan oleh petugas setelah kebakaran terjadi di kilang minyak Pertamina.""",
    """Warga diungsikan akibat kebakaran hebat di area Pertamina Balongan."""
]
```

```python
import pandas as pd
from collections import Counter

# pakai documents kamu sebelumnya
tf = pd.DataFrame(
    [Counter(doc.lower().split()) for doc in documents],
).fillna(0).astype(int)

print("Matriks Term Frequency (TF):")
print(tf)
```

```python
import numpy as np

def length(v):
    return np.sqrt((v ** 2).sum())

def cos_dist(v, w):
    return 1 - (v * w).sum() / (length(v) * length(w))

print("Cosine Distance (Manual):")
print("Doc 0 vs Doc 1 =", cos_dist(tf.loc[0], tf.loc[1]))
print("Doc 0 vs Doc 2 =", cos_dist(tf.loc[0], tf.loc[2]))
```

```python
def cos_sim_manual(v, w):
    return (v * w).sum() / (length(v) * length(w))

print("\nCosine Similarity (Manual):")
print("Doc 0 vs Doc 1 =", cos_sim_manual(tf.loc[0], tf.loc[1]))
print("Doc 0 vs Doc 2 =", cos_sim_manual(tf.loc[0], tf.loc[2]))
```

```python
full_cos_sim_matrix = cosine_similarity(X)

cos_sim_df = pd.DataFrame(
    full_cos_sim_matrix,
    index=[f'Doc {i}' for i in range(len(documents))],
    columns=[f'Doc {i}' for i in range(len(documents))]
)

print("\nFull Cosine Similarity Matrix:")
print(cos_sim_df)
```
